```python
# import all the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# get data
df=pd.read_excel("Honda.xlsx")
df
```

```python
# cleaning
df["Quantity"]=df.Quantity.fillna(df.Quantity.mean())
df
```

```python
df["Unit Price"]=df["Unit Price"].fillna(df["Unit Price"].mean())
df
```

```python
df.isnull().sum()
```

```python
# drop shipped_date
df.drop("Shipped Date",inplace=True,axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Shipper Name"]=df["Shipper Name"].fillna(df["Shipper Name"].mode()[0])
df
```

```python
df.isnull().sum()
```

```python
df["month"]=df["Order_Date"].dt.month
df
```

```python
df["month_name"]=df["Order_Date"].dt.month_name()
df
```

```python
df["day_name"]=df["Order_Date"].dt.day_name()
df
```

```python
df["Quarter"]=df["Order_Date"].dt.quarter
df
```

```python
# set decimal place upto 2 digit
pd.set_option("display.precision",2)
```

```python
df
```

```python
for i in df:
    
    if df[i].dtypes=="object":
        print(i)
        print(df[i].unique())
```

```python
df["sales"]=df["Unit Price"]*df["Quantity"]
df
```

```python
# region wise sales
rs=df[["Region","sales"]].groupby("Region")[["sales"]].sum().sort_values(by="sales",ascending=False)
rs
```

```python
# payment wise sales
ps=df[["Payment Type","sales"]].groupby("Payment Type")[["sales"]].sum().sort_values(by="sales",ascending=False)
ps
```

```python
# product wise sales
prod_sales=df[["Product Name","sales"]].groupby("Product Name")[["sales"]].sum().sort_values(by="sales",ascending=False)[:5]
prod_sales
```

```python
# category wise sales
cat_sales=df[["Category","sales"]].groupby("Category")[["sales"]].sum().sort_values(by="sales",ascending=False)
cat_sales
```

```python
# citywise sales
city_sales=df[["City","sales"]].groupby("City")[["sales"]].sum().sort_values(by="sales",ascending=False)
city_sales
```

```python
# shipper name wise sales
shipper_sales=df[["Shipper Name","sales"]].groupby("Shipper Name")[["sales"]].sum().sort_values(by="sales",ascending=False)
shipper_sales
```

```python
# shipper name wise sales
cust_sales=df[["Customer Name","sales"]].groupby("Customer Name")[["sales"]].sum().sort_values(by="sales",ascending=False)
cust_sales
```

```python
# mont wise sales

month_sales=df[["month","sales"]].groupby("month")[["sales"]].sum()
month_sales
```

```python
quarter_sales=df[["Quarter","sales"]].groupby("Quarter")[["sales"]].sum()
quarter_sales
```
